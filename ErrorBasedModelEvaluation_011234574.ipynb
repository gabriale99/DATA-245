**3. (15%) Chapter 7, exercise 4**

**The use of the kernel trick is key in writing efficient implementations of the support vector machine approach to predictive modelling. The kernel trick is based on the fact that the result of a kernel function applied to a support vector and a query instance is equivalent to the result of calculating the dot product between the support vector and the query instance after a specific set of basis functions have been applied to both—in other words kernel (d, q) = ϕ (d) · ϕ (q).**

---

**a. Using the support vector 〈d[1], d[2]〉 and the query instance 〈q[1], q[2]〉 as examples, show that applying a polynomial kernel with p = 2, kernel(d, q) = (d · q + 1)^2, is equivalent to calculating the dot product of the support vector and query instance after applying the following set of basis functions:**

Chapter 7, exercise 4 a.jpg

<pre>
kernel(d, q) = (d · q + 1)<sup>2</sup>
= (〈d[1], d[2]〉 · 〈q[1], q[2]〉 + 1)<sup>2</sup>
= (d[1]q[1] + q[2]d[2] + 1)<sup>2</sup>
= (d[1]q[1])<sup>2</sup> + (d[2]q[2])<sup>2</sup> + 2(d[1]q[1])(d[2]q[2]) + 2(d[1]q[1]) + 2(d[2]q[2]) + 1

Applying the basis functions to support vectors and queries

ϕ(d) = 〈ϕ<sub>0</sub>(d), ϕ<sub>1</sub>(d), ϕ<sub>2</sub>(d), ϕ<sub>3</sub>(d), ϕ<sub>4</sub>(d), ϕ<sub>5</sub>(d)〉
= 〈d[1]<sup>2</sup>, d[2]<sup>2</sup>, sqrt(2) * d[1] * d[2], sqrt(2) * d[1], sqrt(2) * d[2], 1〉

ϕ(q) = 〈ϕ<sub>0</sub>(q), ϕ<sub>1</sub>(q), ϕ<sub>2</sub>(q), ϕ<sub>3</sub>(q), ϕ<sub>4</sub>(q), ϕ<sub>5</sub>(q)〉
= 〈q[1]<sup>2</sup>, q[2]<sup>2</sup>, sqrt(2) * q[1] * q[2], sqrt(2) * q[1], sqrt(2) * q[2], 1〉

ϕ(d) · ϕ(q) = 〈d[1]<sup>2</sup>, d[2]<sup>2</sup>, sqrt(2) * d[1] * d[2], sqrt(2) * d[1], sqrt(2) * d[2], 1〉 ·
〈q[1]<sup>2</sup>, q[2]<sup>2</sup>, sqrt(2) * q[1] * q[2], sqrt(2) * q[1], sqrt(2) * q[2], 1〉
= d[1]<sup>2</sup>q[1]<sup>2</sup> + d[2]<sup>2</sup>q[2]<sup>2</sup> + (sqrt(2) * d[1] * d[2] * sqrt(2) * q[1] * q[2]) + (sqrt(2) * d[1] * sqrt(2) * q[1]) + (sqrt(2) * d[2] * sqrt(2) * q[2]) + (1 * 1)
= (d[1]q[1])<sup>2</sup> + (d[2]q[2])<sup>2</sup> + 2(d[1]q[1])(d[2]q[2]) + 2(d[1]q[1]) + 2(d[2]q[2]) + 1

This proves the kernel with p = 2 is equal to the the dot product of the support vector and the query instance after applying the basis functions
</pre>

**What is ordinary least square (OLS) method to obtain linear regression model weights?**

Ordinary least squares (OLS) regression obtain the weights by minimizing the error of the prediction.

```python
import numpy as np
import pandas as pd
```

```python
def ols_predict(X, y):
    X = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
    weights = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, y))
    pred = np.dot(X, weights)
    return pred
```

```python
# import file
df = pd.read_csv('/student_scores.csv')
```

```python
X_val = df['Hours'].tolist()
X_val = np.array([[x] for x in X_val])
y_val = df['Scores'].values
```

```python
prediction = ols_predict(X_val, y_val)
result = pd.DataFrame({'Actual': y_val, 'Prediction': prediction})
print(result)
```
